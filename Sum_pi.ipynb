```python
#Khushboo Garg
import numpy as np, matplotlib.pyplot as plt
```

```python
def sum_series(N):
    """
    Compute S(N) without using for loop
    """
    j = np.arange(1, N+1) # j = [1,2, ..., N]
    product = 4 / (2*j - 1) @ (-1)**(j+1) #inner product (dot product) of 2 arrays - Sums the elemental products in one go. Hence, for loop is not needed

    return product
```

```python
#Example
sum_series(3)
```

```python
def plot(N):
    """
    Plot S(N) vs N for 1 <= N <= 50 - computationally show convergence to pi
    """
    if N > 50:
        raise ValueError("N must be <= 50")

    j = np.arange(1, N+1)
    S = np.array([sum_series(k) for k in j])  # calculating S(N) for values of N using for loop
    plt.plot(j, S, marker='o', label="S(N)")
    plt.axhline(np.pi, color='red', linestyle='--', label="Ï€")
    plt.xlabel("N")
    plt.ylabel("S(N)")
    plt.title("Convergence of S(N) to pi")
    plt.legend()
    plt.show()
```

```python
plot(50)
```
