```python
#MULTI DIMENSIONAL ARRAYS
A1 = [b, b+21, b**2] # list of 2D arrays
A1
#A2 = array (b, b+21, b**2) array() takes from 1 to 2 positional arguments but 3 were given -
# the array function is typically used to create an array from a single list or tuple of elements.
A = np.vstack([a, a+21, a**2]) #2D array vertically stacked resulting in a 3*3 array
x = inv(A) @ b # solves A*x=b for x
z = A @ x - b # A(3*3) @ x(3,1) resulting in a (3,1) matrix shape which is same as b, so element vise subtraction
# since x was solved for Ax=b, A@x should be b - which should ideally be 0 (but floating point inaccuracies)
AtA = A.T @ A #symmetric non-negative matrix
z, AtA
```

```python
# example ( b )
err_gr = 0.0
err_05 = 0.0
err_08 = 0.0
err_11 = 0.0
err_20 = 0.0
err_22 = 0.0
err_32 = 0.0
err_35 = 0.0
gr = ( np.sqrt(5) - 1.0)/2.0

for x in range (0 , 100000): # for-loop in Python - the mathematical expression inside abs() should be zero but floating point arithmetic introduces small inaccuracies
    err_gr += abs(x - (1.0/ gr ) * x * gr )
    err_05 += abs(x - (1.0/05.0) * x * 05.0)
    err_08 += abs(x - (1.0/08.0) * x * 08.0)
    err_11 += abs(x - (1.0/11.0) * x * 11.0)
    err_20 += abs(x - (1.0/20.0) * x * 20.0)
    err_22 += abs(x - (1.0/22.0) * x * 22.0)
    err_32 += abs(x - (1.0/32.0) * x * 32.0)
    err_35 += abs(x - (1.0/35.0) * x * 35.0)

print (" err_gr :" , err_gr ) # not zero
print (" err_05 :" , err_05 ) # not zero
print (" err_08 :" , err_08 ) # exactly zero
print (" err_11 :" , err_11 ) # not zero
print (" err_20 :" , err_20 ) # not zero
print (" err_22 :" , err_22 ) # not zero
print (" err_32 :" , err_32 ) # exactly zero
print (" err_35 :" , err_35 ) # not zero
```
