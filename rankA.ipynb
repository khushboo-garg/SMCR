Ques: What rank of matrix means and why it is useful?

Ans: Rank of a matrix is number of linearly independent rows or columns. in other words, number of rows that carry unique information. So it helps determine-


1. If a system of linear equations has solutions.
2. The "usefulness" of rows or columns contributes to the matrix's information.

The rank of the matrix cannot exceed the total number of rows in a matrix, and if a matrix has all rows with zero elements, then the rank of a matrix is said to be zero.

Source:
https://www.geeksforgeeks.org/maths/rank-of-matrix/


```python
import numpy as np
```

```python
m = 2 # number of rows of A
n = 3 # number of columns of A
#A = np.random.randn(m,n) # an m by n matrix of random numbers

row1 = np.random.randn(1, n)
row2 = row1 * 2
A = np.vstack([row1, row2])

AtA = A.T @ A # A−transpose times A . . . n by n
AAt = A @ A.T # A times A−transpose . . . m by m

sizeA = A.shape
rankA = np.linalg.matrix_rank( A )

sizeAtA = AtA.shape
rankAtA = np.linalg.matrix_rank( AtA )

sizeAAt = AAt.shape
rankAAt = np.linalg.matrix_rank( AAt )

print (" size (A ) = " , sizeA )
print (" rank (A ) = " , rankA )
print (" size ( AtA ) =" , sizeAtA )
print (" rank ( AtA ) =" , rankAtA )
print (" size ( AAt ) =" , sizeAAt )
print (" rank ( AAt ) =" , rankAAt )
```

Ques: Why do the rank values make sense?

Answer: ATA and AAT carry the same information about linear independence as A. And so rank(A) = rank(ATA) = rank(AAT).

Ques: For the case of m = 2 , what condition on A would result in the rank of AAT equaling 1?

Ans: Since rank (AAT) = rank (A), this implies rank (A) is also equal to 1. This will happen when there is only one linearly independent/unique row and the other row is just a multiple of other.
